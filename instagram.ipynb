```python
#changing data type to int
data = data.astype({"Followers":"int","Authentic Engagement in M":"int","Engagement Avg. in M":"int"})
data.head()
```

```python
top20followers = data.nlargest(20,'Followers')
```

```python
#bargraph
from matplotlib import pyplot as plt

plt.figure(figsize=(12,8))
fig2= plt.bar(top20followers.Category, top20followers.Followers,label="Followers in Million",width=0.8, color='pink')
plt.xticks(fontsize=12, rotation=90)
plt.ylabel("Followers in Million", fontsize=16)
plt.xlabel("Category", fontsize=16)
plt.title("Top 20 Catgories by Followers", fontsize=20)
plt.tight_layout()
plt.ticklabel_format(axis="y", style='plain')
```

```python
#showing categories using pie chart
types= data.Category.value_counts()
types.reset_index()
```

```python
#creatong pie-chart

plt.pie(types.values,labels=types.index, labeldistance=1.2,autopct='%1.2f%%',shadow = True, startangle=60)
plt.title("Different categories", fontsize=20)
plt.plot()
```
